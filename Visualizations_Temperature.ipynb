```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('query-result.csv')

# Extract the 'temp1' and 'time' columns from the data
temp1_values = data['temp1']
time_values = data['time']

# Convert time_values to datetime if it's not already in datetime format
time_values = pd.to_datetime(time_values)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(time_values, temp1_values)

# Set the labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Temperature')
ax.set_title('Temperature Variation')

# Rotate and align the x-axis tick labels
plt.xticks(rotation=45, ha='right')

#Save the plot
plt.savefig('plot_image.png')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('query-result.csv')

# Calculate the number of data points to plot (one-fourth of the total)
num_points = len(data) // 4

# Extract the 'temp1' and 'time' columns from the data
temp1_values = data['temp1'][:num_points]
time_values = pd.to_datetime(data['time'][:num_points])

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data with a red line
ax.plot(time_values, temp1_values, linewidth=2)

# Set the labels and title with larger font size
ax.set_xlabel('Time', fontsize=14)
ax.set_ylabel('Temperature', fontsize=14)
ax.set_title('Temperature Variation (Monthly)', fontsize=16)

# Rotate and align the x-axis tick labels
plt.xticks(rotation=45, ha='right')

# Add grids to the graph background
ax.grid(True, linestyle='--', alpha=0.7)

# Adjust the layout to prevent cutting off the bottom part
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)  # Increase bottom padding

# Save the plot as an image
plt.savefig('plot_image2.png')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Read the CSV file
data = pd.read_csv('query-result.csv')

# Calculate the number of data points to plot (one-fourth of the total)
num_points = len(data) // 200

# Extract the 'temp1' and 'time' columns from the data
temp1_values = data['temp1'][:num_points]
time_values = pd.to_datetime(data['time'][:num_points])

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data with a red line
ax.plot(time_values, temp1_values, linewidth=2)

# Set the labels and title with larger font size
ax.set_xlabel('Time', fontsize=14)
ax.set_ylabel('Temperature', fontsize=14)
ax.set_title('Temperature Variation', fontsize=16)

# Format x-axis tick labels to show only the time
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

# Rotate and align the x-axis tick labels
plt.xticks(rotation=45, ha='right')

# Add grids to the graph background
ax.grid(True, linestyle='--', alpha=0.7)

# Adjust the layout to prevent cutting off the bottom part
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)  # Increase bottom padding

# Save the plot as an image
plt.savefig('plot_image3.png')

# Display the plot
plt.show()
```
